```python
from ultralytics import YOLO


model = YOLO("yolov8s.pt")  # kamu bisa ganti ke yolov8s.pt, yolov8m.pt, dst

model.train(
    data="/content/skin_cancer-2/data.yaml",  # path dari Roboflow
    epochs=10,
    imgsz=640,
    batch=16,
    name="yolo_skin_cancer"
)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load an official model
model = YOLO("/content/runs/detect/yolo_skin_cancer/weights/best.pt")  # load a custom model

# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map  # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps  # a list contains map50-95 of each category
```
